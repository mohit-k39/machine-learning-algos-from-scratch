```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

n = 1000
d = 5

X = 2 * np.random.rand(n, d)

true_theta = np.array([4, 3, -2, 1.5, 0.5])
# true_theta = np.array([2])

noise = np.random.randn(n)

y = X @ true_theta + noise

print(X.shape)
print(y.shape)
```

```python
class LinearRegressionGradiendDescent():

    def fit(self, X, y):
        theta = np.zeros(X.shape[1])
        for i in range(1000):
            loss = (theta@X.T-y)@(theta@X.T-y)
            loss_gradient = X.T@X@theta-X.T@y
            theta = theta - 0.00001*loss_gradient
            new_loss = (theta@X.T-y)@(theta@X.T-y)
            if abs(loss - new_loss) <= 0.001:
                print(i)
                break
        return theta

class LinearRegressionClosedForm():

    def fit(self, X, y):
        theta = np.linalg.inv(X.T@X)@X.T@y
        return theta
```

```python
lr = LinearRegressionGradiendDescent()
lrcf = LinearRegressionClosedForm()
theta_learned = lr.fit(X[:1000],y[:1000])
theta_learned_cf = lrcf.fit(X[:1000],y[:1000])
```

```python
print(f"gradient descent solution : {theta_learned}")
print(f"closed fom solution : {theta_learned_cf}")
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept=False)

model.fit(X, y)

print("Gradient Descent theta:")
print(theta_learned)

print("\nclosed form theta:")
print(theta_learned_cf)

print("\nSklearn theta:")
print(model.coef_)
```
